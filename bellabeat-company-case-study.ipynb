**Determining the credibility of data:**
The data set satisfies most of ROCCC, which are 5 criteria to assess whether data is good or bad according to the Google Data Analytics course (Reliable, Original, Comprehensive, Current and Cited). 
- Reliable: The data was generated by FitBit users and collected via Amazon Mechanical Turk, a reputable platform for data collection. However, the sample of this data set is only 33, which may not be enough to be representative of the population and may lead to bias.
- Original: The data was collected directly from the FitBit devices used by the participants.
- Comprehensive: The data contains a lot of information needed to analyze and find insights about smart device wearers.
- Current: The data set was collected between 03.12.2016-05.12.2016, which means it is more than 6 years old and is not so new anymore.
- Cited: The source of the data and information about its authors are provided.

To sum up, although the data set is not new and its sample is not very large, it comes from reliable sources and can provide many useful information for the analysis of device users habits.

# IV. Process:

In this step we will import, clean and transform data to make it ready to be analyzed.

**1. Installing necessary packages**


```r
library(tidyverse) 
library(here) 
library(skimr) 
library(janitor) 
library(dplyr) 
library(lubridate) 
library(ggplot2) 
library(readr) 
```

**2. Importing data sets**

In this case study, we will focus on three files:
- daily_Activity
- hourly_Intensity
- sleep_Day

```r
daily_activity <- read_csv("/kaggle/input/fitbit/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv") 
hourly_intensity <- read_csv("/kaggle/input/fitbit/Fitabase Data 4.12.16-5.12.16/hourlyIntensities_merged.csv") 
sleep_day <- read_csv("/kaggle/input/fitbit/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv") 
```

**3. Previewing data sets**

```r
glimpse(daily_activity) 
glimpse(hourly_intensity) 
glimpse(sleep_day)
```

**4. Cleaning and formatting data**

**a. Cleaning column names:**

We will use clean_names function to replaces spaces, special characters, and uppercase letters with underscores and lowercase letters, which makes the column names easier to work with in R.

```r
daily_activity <- clean_names(daily_activity) 
hourly_intensity <- clean_names(hourly_intensity) 
sleep_day <- clean_names(sleep_day) 
```

**c. Converting columns containing time to date or datetime formats**

```r
daily_activity$activity_date <- as.Date(daily_activity$activity_date, format = "%m/%d/%Y") 

hourly_intensity <- hourly_intensity %>% 
  mutate(activity_hour = as.POSIXct(activity_hour, format = "%m/%d/%Y %I:%M:%S %p")) 

sleep_day <- sleep_day %>% 
  mutate(sleep_day = as.Date(sleep_day, format = "%m/%d/%Y %I:%M:%S %p"))
```

**d. Removing duplicates**

```r
daily_activity <- distinct(daily_activity, .keep_all = TRUE) 
hourly_intensity <- distinct(hourly_intensity, .keep_all = TRUE) 
sleep_day <- distinct(sleep_day, .keep_all = TRUE) 
```

After removing, we will check if there are any duplicates left

```r
sum(duplicated(daily_activity)) 
sum(duplicated(hourly_intensity)) 
sum(duplicated(sleep_day)) 
```

**e. Verifying the number of users in each data frame**

```r
n_unique(daily_activity$id) 
n_unique(hourly_intensity$id) 
n_unique(sleep_day$id) 
```

**f. Merging data frames** 

In this project, we are going to analyse the relationship between sleep patterns of users with their daily activity data. Therefore, we need to merge two data frames daily_activity and sleep_day together. First, we need to change the name of "activity_date" column in daily_activity frame and "sleep_day" column in sleep_day frame to "date" before merging as these two columns need to have the same name.

```r
names(daily_activity)[names(daily_activity) == "activity_date"] <- "date" 
names(sleep_day)[names(sleep_day) == "sleep_day"] <- "date"
```

Then we will merge these 2 frames

```r
daily_activity_sleep <- merge(daily_activity, sleep_day, by=c ("id", "date")) 
```

```r
head(daily_activity_sleep)
```

While conducting this project, many researchers have investigated the correlation between people's health habits and their sleep quality by using "total_minutes_asleep" as a variable. However, in my opinion, a person's total sleep duration is largely influenced by their daily routine, including their bedtime and wake-up time. Therefore, it may be unreasonable to use "total_minutes_asleep" as an indicator of sleep quality in this case.

Numerous studies have suggested that physical activity can have an impact on the time it takes a person to fall asleep. Hence, in this project, it would be more appropriate to analyze the relationship between a person's health habits and the time it takes them to fall asleep instead. This duration can be calculated by subtracting the total minutes asleep from the total time in bed. To do this, we will add a new column titled "time_to_fall_asleep" in the merged dataset.

```r
daily_activity_sleep <- daily_activity_sleep %>% 
  mutate(time_to_fall_asleep = total_time_in_bed - total_minutes_asleep) 
```

```r
head(daily_activity_sleep)
```

# V. Analyze and Share:

**1. Summary**

First we need to get a summary of the data sets to have some overviews

```r
summary(daily_activity_sleep)
```

From the summary, we can notice many issues related to the health habits of smart device users:

- Users walk 8,515 steps/day on average, which can be considered "somewhat active" (Tudor-Locke & Bassett, 2004). However, this needs to be increased, as the CDC recommends people to aim for at least 10,000 steps a day for general health (Huizen, 2021).

**2. Correlation**

Next, it is necessary to discover the relationship between variables. In this case study, we will determine the relationships between:
- Total steps and calories burned
- Total steps and time to fall asleep
- Calories burned and time to fall asleep
- Each of the 4 types of active minutes and time to fall asleep

**a. Total steps and calories burned**

```r
ggplot(daily_activity, aes(x = total_steps, y = calories)) + 

  geom_jitter() + 

  geom_point() + 

  geom_smooth(method = "lm", se = FALSE) + 

  labs(title = "Total Steps vs Calories burned", x = "Total steps", y = "Calories") 
cor(daily_activity$total_steps, daily_activity$calories)
```

Although it is obvious that we will burn more calories when taking more steps, we can see that the correlation between the two variables in this dataset is not strong. This could be due to the small sample size, as mentioned earlier. Another reason could be that steps are not the only factor affecting calories burned. There are many other factors that can be correlated with calories burned, such as age, body composition, body size, and diet, which need to be taken into consideration in further studies.

Moving forward, we will investigate whether calories burned or total steps are correlated with the time users fall asleep.

```r
ggplot(daily_activity_sleep, aes(x = total_steps, y = time_to_fall_asleep)) + 

  geom_jitter() + 

  geom_point() + 

  geom_smooth(method = "lm", se = FALSE) + 

  labs(title = "Total Steps vs Time to fall Asleep", x = "Total steps", y = "Time to fall asleep") 
cor(daily_activity_sleep$total_steps, daily_activity_sleep$time_to_fall_asleep)
```

The time users fall asleep has no correlation with number of steps they take. Now we will explore its relationship with the total calories burned a day.

```r
ggplot(daily_activity_sleep, aes(x = calories, y = time_to_fall_asleep)) + 

  geom_jitter() + 

  geom_point() + 

  geom_smooth(method = "lm", se = FALSE) + 

  labs(title = "Calories vs times to fall asleep", x = "Calories burned", y = "Time to fall asleep") 
cor(daily_activity_sleep$calories, daily_activity_sleep$time_to_fall_asleep) 
```

From the results, we can observe a weak negative correlation between calories burned and time to fall asleep. This means that people tend to fall asleep faster if they burn more calories. Although this relationship is weak, it encourages further studies to explore the factors that affect sleep quality, including more dependent variables such as diet, sleep schedule, stress, and medical conditions.

Next, we will investigate if the number of active minutes in a day has an impact on the time it takes users to fall asleep.

```r
cor(daily_activity_sleep$very_active_minutes, daily_activity_sleep$time_to_fall_asleep)
cor(daily_activity_sleep$fairly_active_minutes, daily_activity_sleep$time_to_fall_asleep)
cor(daily_activity_sleep$lightly_active_minutes, daily_activity_sleep$time_to_fall_asleep)
cor(daily_activity_sleep$sedentary_minutes, daily_activity_sleep$time_to_fall_asleep)
```

```r
ggplot(daily_activity_sleep, aes(x = fairly_active_minutes, y = time_to_fall_asleep)) + 

  geom_jitter() + 

  geom_point() + 

  geom_smooth(method = "lm", se = FALSE) + 

  labs(title = "Fairly active minutes vs times to fall asleep", x = "Fairly active minutes", y = "Time to fall asleep") 
 
```

We can see that the number of very active minutes, lightly active minutes and sedentary minutes is not correlated with how easy people fall asleep. However, the time to fall asleep has a positive association with the number of fairly active minutes. This means that the more time people spend for fairly active activities, the faster they can fall asleep at night. Because of this, encouraging people to take part in fairly active activities may improve their sleep quality.

**3. Trends in daily activity and health habits of users**

**a. Classifying people by active levels**

We will categorize users based on the number of steps taken a day to see what percentage of smart device users each group account for. Classification is made based on Tudor' research (cited in Bumgardner, 2020):

- Sedentary: Less than 5,000 steps/day
- Low Active: 5,000 to 7,499 steps/day
- Somewhat Active: 7,500 to 9,999 steps/day
- Active: 10,000 to 12,500 steps/day
- Very Active: More than 12,500 steps/day

First, we need to calculate each users' average number of steps per day:

```r
average_step <- daily_activity_sleep %>% 

  group_by(id) %>% 

  summarise(avgsteps = mean(total_steps)) 
```

```r
head(average_step)
```

Then we classify users based on their average steps

```r
average_step <- average_step %>% 

  mutate(classification = case_when( 

    avgsteps < 5000 ~ "Sedentary", 

    avgsteps >= 5000 & avgsteps < 7500 ~ "Low Active", 

    avgsteps >= 7500 & avgsteps < 10000 ~ "Somewhat Active", 

    avgsteps >= 10000 & avgsteps <= 12500 ~ "Active", 

    avgsteps > 12500 ~ "Very Active" 

  )) 
```

Next, we calculate the percentage each group accounts for

```r
classification_percent <- average_step %>% 

  group_by(classification) %>% 

  summarise(Count = n()) %>% 

  mutate(Percentage = round(Count/sum(Count)*100, 2)) 
```

The order level of groups needs to be set

```r
order_levels <- c("Sedentary", "Low Active", "Somewhat Active", "Active", "Very Active") 

classification_percent$classification <- factor(classification_percent$classification, levels = order_levels) 
```

Finally, we create a pie chart to demonstrate the distribution of users types based on average steps

```r
pie_chart <- ggplot(classification_percent, aes(x = "", y = Percentage, fill = classification)) + 

  geom_bar(stat = "identity", width = 1) + 

  coord_polar("y", start = 0) + 

  ggtitle("Distribution of User Types") + 

  theme_void() + 

  geom_text(aes(label = paste0(Percentage, "%")), position = position_stack(vjust = 0.5), size = 4) + 
  
  theme(plot.title = element_text(hjust = 0.5, face = "bold"))
```

```r
pie_chart
```

The graph indicates that the majority (37.5%) of smart device users are somewhat active, which is a positive trend. However, the percentages of people with sedentary and low active lifestyles are very high, both at nearly 21%, and need to be decreased. In contrast, the proportions of active and very active users are very low, at only 16.67% and 4.17%, respectively, and need to be improved.

**b. Distribution of intensity in a day**

Before drawing a graph to demonstrate users' average intensity in 24 hours, we need to split the "activity_hour" column in "hour_intensity" data frame into "date" and "time"

```r
hourly_intensity <- hourly_intensity %>% 

  separate(activity_hour, into = c("date", "time"), sep= " ") %>% 

  mutate(date = ymd(date))  
```

We will now draw a histogram to show hourly intensity distribution

```r
hourly_intensity %>% 

  group_by(time) %>% 

  summarize(average_intensity = mean(total_intensity)) %>% 

  ggplot() + 

  geom_col(mapping = aes(x=time, y = average_intensity, fill = average_intensity)) +  

  labs(title = "Intensity throughout the day", x="", y="") +  

  scale_fill_gradient(low = "green", high = "red")+ 

  theme(axis.text.x = element_text(angle = 90)) 
```

We can see that 5pm to 7pm are the time when people are most active. This could be because most of them choose to do exercise after coming home from schools or workplaces. This can be the suitable time for Bellabeat app to send messages to users reminding them to do exercise.

**4. Trends in time users wear devices**

As Bellabeat app is connected with Bellabeat wealness watch, finding ways to make people more engaged with the tracker can help the company increase as well as retain app users. We will now discover some trends and patterns in terms of the time people use smart devices. The time an user wear the tracker in a day is calculated by the total of minutes recorded. 

**a. Distribution of wearing times**

Before drawing the graph, we need to calculate the average wearing time of each users and group them by user ID.  

```r
avg_wearing_time <- daily_activity %>% 

  mutate(time_worn = very_active_minutes + fairly_active_minutes + lightly_active_minutes + sedentary_minutes) %>% 

  group_by(id) %>% 

  summarise(avg_time_worn = mean(time_worn))
```

We can now visualize the wearing time distribution

```r
ggplot(data = avg_wearing_time, aes(x = avg_time_worn)) + 

  geom_histogram(binwidth = 30, fill = "dodgerblue", color = "black") + 

  labs(title = "Distribution of Time Worn", x = "Time Worn (minutes)", y = "Frequency") + 

  scale_x_continuous(breaks = seq(0, 1440, by = 60), labels = seq(0, 1440, by = 60)) 
```

From the graph, it is clear that most people wear the tracker more than 1,380 minutes (23 hours) a day. As they have the tendency to wear the device almost all day, they may look for features such as long battery life or comfortable straps when choosing trackers. Focus on these features and promoting them can attract more customers, which lead to more app users. 

Besides,as the analysis shows that users wear the device for the majority of the day, Bellabeat may want to emphasize the device's ability to track daily activity and provide personalized insights for a healthy lifestyle. This means that they may want to focus their marketing efforts on highlighting the device's ability to track steps, calories burned, and other metrics throughout the day. They could also emphasize the device's ability to provide personalized insights and recommendations based on the user's data, which can help users adopt healthier habits and reach their fitness goals.

**b. Average wearing time over the period**

Let's see how the average time users wear the tracker change over time.

```r
daily_activity %>%
  mutate(time_worn = very_active_minutes + fairly_active_minutes + lightly_active_minutes + sedentary_minutes,
         date = as.Date(date)) %>%
  group_by(date) %>%
  summarize(avg_time_worn = mean(time_worn)) %>%
  ggplot(aes(x = date, y = avg_time_worn)) +
  geom_line(color = "dodgerblue") +
  labs(x = "Date", y = "Average time worn (minutes)", title = "Average wearing time over time") +
  scale_x_date(date_breaks = "1 day", date_labels = "%m/%d/%y") +
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5))
```

As the line graph indicates a decrease in average wearing time over time, Bellabeat may want to investigate the reasons behind this trend and come up with more initiatives to keep users engaged.
They may want to target marketing efforts towards users who have stopped wearing the device or encourage more consistent use through rewards or incentives.

# VI. Act (Recommendations)

Based on the key insights mentioned above, some recommendations for the marketing strategy of the Bellabeat app are proposed:

1. Further studies: Regarding the limitations of the dataset used and the research gaps identified above, Bellabeat should use their own tracking data for better analysis in future studies. This should include a larger sample size and more information, such as age, nutrition data, exercise data, and body measurements.

2. Daily notifications and blog posts: Bellabeat could send daily in-app notifications to encourage users to take at least 10,000 steps every day and have at least 22 fairly active minutes a day. Additionally, the app could send workout reminders from 5 pm to 7 pm, especially on days that people tend to workout less. Alarms will also be sent to users if they have too many sedentary minutes or take too few steps in a day. 

   Besides, in the app, there should be a blog section where insightful posts about health and fitness will be uploaded, and users can access them to read these posts. Bellabeat can also choose featured blog posts and send them to users every day based on their preferences. For example, if users care more about diet, then more posts about nutrition and diet knowledge will be recommended to them.

3. Emphasize the importance of sleep: Since many device users do not sleep enough, Bellabeat could create content that emphasizes the importance of getting enough sleep for overall health and wellness. This could include blog posts and social media campaigns. They should also develop a sleep tracker feature in their app, which can give users insights about their sleep quality, and the promotion of this feature would be integrated into content about the importance of sleep.

4. Sleep-inducing features: To help users fall asleep easier, Bellabeat can introduce new features, such as bedtime reminders, to help users establish a regular bedtime routine. Guided meditation and relaxation exercises can also be uploaded on the app to help users reduce stress and anxiety, which can improve their ability to fall asleep and stay asleep.

5. Personalized recommendations: Bellabeat could provide personalized workout and nutrition recommendations to users based on their fitness goals and health conditions.

6. Promoting fitness tracker: Bellabeat should invest in developing device features such as long battery life, waterproof design, and comfortable straps to attract customers and retain them in the long run. Additionally, the company should emphasize the device's ability to track daily activity and offer personalized insights for a healthy lifestyle.

7. Rewards and incentives:

   a. Gamification: Bellabeat could use gamification to encourage consistent use of their device, such as offering rewards or achievements for reaching certain milestones or completing certain activities.

   b. Social sharing: Bellabeat could encourage users to share their activity data and progress with friends and family on social media. This could help create a sense of community and accountability, which may encourage users to wear the device more consistently.
